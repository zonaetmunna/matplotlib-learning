```python
# Color Maps

import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Choose a colormap
cmap = plt.get_cmap("viridis")  # Example colormap

# Create a line plot with the chosen colormap
plt.plot(x, y, color=cmap(0.6), label="Sine Wave")

# Add labels and a title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot with Custom Color")

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Styles

import matplotlib.pyplot as plt

# Choose a style
plt.style.use("ggplot")  # Example style

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a scatter plot with the chosen style
plt.scatter(x, y, marker="o", color="blue")

# Add labels and a title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot with Custom Style")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Choose a colormap
cmap = plt.get_cmap("viridis")

# Choose a style
plt.style.use("seaborn-darkgrid")

# Create a line plot with the chosen colormap and style
plt.plot(x, y, color=cmap(0.6), label="Sine Wave")

# Add labels and a title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot with Custom Color and Style")

# Display the legend
plt.legend()

# Show the plot
plt.show()
```
