```python
# Creating a Simple Scatter Plot

import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a simple scatter plot
plt.scatter(x, y)

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of X and Y')

# Display the plot
plt.show()
```

```python
# Customizing Scatter Plots

import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a simple scatter plot
# Custom marker style Square markers
# Custom marker color
# Custom marker size
plt.scatter(x, y,marker='s',color='red',s=100) 

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of X and Y')

# Display the plot
plt.show()
```
