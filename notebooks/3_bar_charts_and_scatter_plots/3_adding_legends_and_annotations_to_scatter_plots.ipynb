```python
# Creating Scatter Plots with Legends

import matplotlib.pyplot as plt

# Sample data for two groups
x1 = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
x2 = [1, 2, 3, 4, 5]
y2 = [1, 3, 5, 7, 9]

# Create scatter plots for each group with labels
plt.scatter(x1, y1, label='Group 1', marker='o', color='blue')
plt.scatter(x2, y2, label='Group 2', marker='x', color='red')

# Add labels and a legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Legends')
plt.legend()

# Display the plot
plt.show()
```

```python
# Adding Annotations to Scatter Plots

import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a scatter plot
plt.scatter(x, y, marker='o', color='blue')

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Annotations')

# Add an annotation for a specific data point
plt.annotate('Point (3, 6)', xy=(3, 6), xytext=(4, 7),
             arrowprops=dict(arrowstyle='->', color='red'))

# Display the plot
plt.show()
```

```python
# Exercise: Create a Bar Chart and a Scatter Plot with Meaningful Data

import matplotlib.pyplot as plt

# Sample data
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [350, 460, 280, 590]

# Create a bar chart
plt.bar(products, sales)

# Add labels and a title
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Product Sales Bar Chart')

# Display the bar chart
plt.show()


# Sample data
hours_studied = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores = [60, 65, 70, 75, 80, 85, 90, 95]

# Create a scatter plot
plt.scatter(hours_studied, exam_scores, marker='o', color='blue')

# Add labels and a title
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Scatter Plot of Hours Studied vs. Exam Scores')

# Display the scatter plot
plt.show()
```
