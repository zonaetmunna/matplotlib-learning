```python
# Creating a Simple Bar Chart
import matplotlib.pyplot as plt

# Sample data
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [350, 460, 280, 590]

# Create a simple bar chart
plt.bar(products, sales)

# Add labels and a title
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Product Sales Bar Chart')

# Display the plot
plt.show()
```

```python
# Customizing Bar Charts
import matplotlib.pyplot as plt

# Sample data
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [350, 460, 280, 590]


# Custom colors, Custom bar width for bars
plt.bar(products, sales, color=['blue', 'green', 'orange', 'red'])
# Add labels to the bars
for i, sale in enumerate(sales):
    plt.text(products[i], sale + 10, str(sale), ha='center')

plt.show()
```

```python
# Stacked Bar Chart
import numpy as np

# Sample data
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales1 = [350, 460, 280, 590]
sales2 = [220, 380, 420, 310]

# Calculate cumulative sales for a stacked bar chart
cumulative_sales1 = np.cumsum(sales1)
cumulative_sales2 = np.cumsum(sales2)

# Create a stacked bar chart
plt.bar(products, sales1, label='Quarter 1', color='blue')
plt.bar(products, sales2, label='Quarter 2', bottom=sales1, color='green')

# Add labels and a legend
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Stacked Bar Chart of Product Sales')
plt.legend()
```
