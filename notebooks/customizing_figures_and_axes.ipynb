```python
import matplotlib.pyplot as plt

# Create a figure and a set of axes
fig, ax = plt.subplots()
# Data for the line plot
x = [1, 2, 3, 4, 5]
y = [10, 16, 12, 18, 14]

# Add a line plot to the axes with customizations
ax.plot(x, y, linestyle='--', marker='o', color='blue', label='Data')
ax.set_title('Simple Line Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.grid(True)
ax.legend()

# Display the plot
plt.show()
```

```python
# Create Your First Matplotlib Plot (Simple Line Plot)
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)  # Create an array of 100 points from 0 to 10
y = np.sin(x)  # Compute the sine of each point

# Create the plot
plt.plot(x, y)

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

# Display the plot
plt.show()


```
