```python
# Customizing Titles and Labels
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot
plt.plot(x, y)

# Add a title
plt.title('Customized Line Plot')

# Add labels to the axes
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Display the plot
plt.show()
```

```python
# Adding Marker Symbols

import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot with markers
plt.plot(x, y, marker='o')

# Add a title and labels
plt.title('Customized Line Plot')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Display the plot
plt.show()
```

```python
# Customizing Line Styles and Colors

import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot with custom style and color
plt.plot(x, y, linestyle='--', color='red')

# Add a title and labels
plt.title('Customized Line Plot')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Display the plot
plt.show()
```
