```python
# Creating Multiple Plots
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

# Create a 1x2 grid of plots (1 row, 2 columns)
plt.subplot(1, 2, 1)  # The first plot
plt.plot(x, y1)
plt.title('Plot 1')

plt.subplot(1, 2, 2)  # The second plot
plt.plot(x, y2)
plt.title('Plot 2')

# Adjust spacing between plots
plt.tight_layout()

# Display the figure
plt.show()
```

```python
# Customizing Multiple Plots

import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

# Create a 1x2 grid of plots
plt.subplot(1, 2, 1)
plt.plot(x, y1, linestyle='--', color='blue')
plt.title('Line Plot 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.subplot(1, 2, 2)
plt.plot(x, y2, linestyle='-.', color='green')
plt.title('Line Plot 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.tight_layout()
plt.show()
```
