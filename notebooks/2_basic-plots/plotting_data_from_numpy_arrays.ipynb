```python
# Plotting Data from NumPy Arrays

import numpy as np
import matplotlib.pyplot as plt

# Create NumPy arrays for X and Y values
x = np.arange(0, 10, 0.1)
y = np.sin(x)

# Create a line plot using the NumPy arrays
plt.plot(x, y)

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot of sin(x)')

# Display the plot
plt.show()
```

```python
# Customizing Plots with NumPy Data

import numpy as np
import matplotlib.pyplot as plt

# Create NumPy arrays for X and Y values
x = np.arange(0, 10, 0.1)
y = np.sin(x)

# Create a customized line plot
plt.plot(x, y, linestyle='--', color='blue', marker='o')

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Line Plot of sin(x)')

# Set custom axis limits
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)

# Display the plot
plt.show()
```
