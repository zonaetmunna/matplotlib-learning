```python
# Creating a Combined Visualization

import matplotlib.pyplot as plt

# Sample exam scores
exam_scores = [60, 70, 75, 80, 85, 90, 92, 95, 98, 100]

# Create a figure with two subplots (1 row, 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Plot a histogram in the first subplot
axs[0].hist(exam_scores, bins=5, edgecolor='black', color='blue')
axs[0].set_xlabel('Exam Scores')
axs[0].set_ylabel('Frequency')
axs[0].set_title('Histogram of Exam Scores')

# Plot a box plot in the second subplot
axs[1].boxplot(exam_scores, vert=False, boxprops={'color': 'blue'}, whiskerprops={'color': 'green'})
axs[1].set_ylabel('Exam Scores')
axs[1].set_title('Box Plot of Exam Scores')

# Adjust spacing between subplots
plt.tight_layout()

# Display the combined visualization
plt.show()
```
