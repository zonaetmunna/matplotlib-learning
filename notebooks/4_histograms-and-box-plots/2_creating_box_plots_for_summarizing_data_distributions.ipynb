```python
# Creating a Simple Box Plot

import matplotlib.pyplot as plt

# Sample exam scores
exam_scores = [60, 70, 75, 80, 85, 90, 92, 95, 98, 100]

# Create a box plot
plt.boxplot(exam_scores)

# Add labels and a title
plt.ylabel('Exam Scores')
plt.title('Box Plot of Exam Scores')

# Display the box plot
plt.show()
```

```python
# Customizing Box Plots

import matplotlib.pyplot as plt

# Sample exam scores
exam_scores = [60, 70, 75, 80, 85, 90, 92, 95, 98, 100]

# Create a box plot
# Custom Box and Whisker Colors
# Notched box plot
plt.boxplot(exam_scores, boxprops={'color': 'blue'}, whiskerprops={'color': 'green'},vert=False,notch=True)

# Add labels and a title
plt.ylabel('Exam Scores')
plt.title('Box Plot of Exam Scores')

# Display the box plot
plt.show()
```

```python
# Multiple Box Plots

import matplotlib.pyplot as plt

# Multiple box plots
data1 = [60, 70, 75, 80, 85]
data2 = [90, 92, 95, 98, 100]
plt.boxplot([data1, data2], labels=['Group 1', 'Group 2'])

# Add labels and a title
plt.ylabel('Exam Scores')
plt.title('Box Plot of Exam Scores')

# Display the box plot
plt.show()
```
