```python
# Creating a Simple Histogram

import matplotlib.pyplot as plt

# Sample exam scores
exam_scores = [60, 70, 75, 80, 85, 90, 92, 95, 98, 100, 100, 100]

# Create a histogram
plt.hist(exam_scores, bins=5, edgecolor='black')

# Add labels and a title
plt.xlabel('Exam Scores')
plt.ylabel('Frequency')
plt.title('Histogram of Exam Scores')

# Display the histogram
plt.show()
```

```python
# Customizing Histograms

import matplotlib.pyplot as plt

# Sample exam scores
exam_scores = [60, 70, 75, 80, 85, 90, 92, 95, 98, 100, 100, 100]

# Create a histogram
# Density Plot
# Custom Colors
# Adding More Bins
# Adding a Cumulative Histogram
plt.hist(exam_scores, bins=10, edgecolor='black', color='blue', density=True,cumulative=True)

# Add labels and a title
plt.xlabel('Exam Scores')
plt.ylabel('Frequency')
plt.title('Histogram of Exam Scores')

# Display the histogram
plt.show()
```
